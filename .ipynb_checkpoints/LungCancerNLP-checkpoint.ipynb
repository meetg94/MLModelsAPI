```python
# Categorize the data into genetic, gender, and environmental factors
genetic_columns = ['age', 'genetic_risk', 'chronic_lung_disease']
gender_columns = ['gender']
environmental_columns = ['air_pollution', 'alcohol_use', 'dust_allergy',
                         'occupational_hazards', 'chronic_lung_disease',
                         'balanced_diet', 'obesity', 'smoking',
                         'passive_smoker', 'chest_pain', 'coughing_of_blood',
                         'fatigue', 'weight_loss', 'shortness_of_breath',
                         'wheezing', 'swallowing_difficulty',
                         'clubbing_of_finger_nails', 'frequent_cold',
                         'dry_cough', 'snoring']

# Subset the data based on the categories
genetic_data = data[genetic_columns]
gender_data = data[gender_columns]
environmental_data = data[environmental_columns]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# First we need to make sure 'age' is treated as a numeric column. If it's not, convert it.
data['age'] = pd.to_numeric(data['age'], errors='coerce')

# Compute the average age for each level
average_age_by_level = data.groupby('level_encoded')['age'].mean()

# Reset the index to convert the index into a column
average_age_by_level = average_age_by_level.reset_index()

# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x='level_encoded', y='age', data=average_age_by_level)
plt.title('Average age by level')
plt.xlabel('level_encoded')
plt.ylabel('Average age')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Convert all categorical variables to numeric format
for column in data.columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')

# Define the features and target
features = ['air_pollution', 'alcohol_use', 'dust_allergy', 'occupational_hazards',
            'chronic_lung_disease', 'balanced_diet', 'obesity', 'smoking',
            'passive_smoker', 'chest_pain', 'coughing_of_blood', 'fatigue',
            'weight_loss', 'shortness_of_breath', 'wheezing',
            'swallowing_difficulty', 'clubbing_of_finger_nails', 'frequent_cold',
            'dry_cough', 'snoring']

target = 'level_encoded'

# Initialize the model
model = RandomForestClassifier(n_estimators=100, random_state=0)

# Fit the model
model.fit(data[features], data[target])

# Get the feature importances
importances = model.feature_importances_

# Convert the importances into a pandas DataFrame
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

# Sort the DataFrame by the importance
importance_df_RFC = importance_df.sort_values(by='Importance', ascending=False)
```

```python
import xgboost 
from xgboost import XGBClassifier

# Initialize the model
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# Fit the model
model.fit(data[features], data[target])

# Get feature importances
importances = model.feature_importances_

# Create a DataFrame of feature importances
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

# Sort DataFrame by the importance column
importance_df_xGBC = importance_df.sort_values(by='Importance', ascending=False)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the model
model = GradientBoostingClassifier(random_state=0)

# Fit the model
model.fit(data[features], data[target])

# Get feature importances
importances = model.feature_importances_

# Create a DataFrame of feature importances
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importances
})

# Sort DataFrame by the importance column
importance_df_GBC = importance_df.sort_values(by='Importance', ascending=False)
```

```python
# First, let's merge all three dataframes on 'Feature'
merged_df = pd.merge(importance_df_RFC, importance_df_xGBC, on='Feature', suffixes=('_RFC', '_xGBC'))
merged_df = pd.merge(merged_df, importance_df_GBC, on='Feature')
merged_df.rename(columns={'Importance': 'Importance_GBC'}, inplace=True)

# Now, let's compute the mean importance across all models
merged_df['Mean_Importance'] = merged_df[['Importance_RFC', 'Importance_xGBC', 'Importance_GBC']].mean(axis=1)

# Sort the DataFrame by the mean importance
merged_df.sort_values(by='Mean_Importance', ascending=False, inplace=True)
```

```python
# Get the names of the top features
top_features = merged_df.head(15)['Feature'].tolist()

print(top_features)
```

```python
selected_columns = top_features + gender_columns + genetic_columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
import pickle

# Choose your features and target
X = data[selected_columns]  # Put your selected feature names here
y = data['level_encoded']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the model
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train, y_train)

# Save the trained model to a file
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Make predictions
y_pred = model.predict(X_test)

# Measure accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
